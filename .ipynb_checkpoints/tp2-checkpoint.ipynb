```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import warnings

warnings.simplefilter('ignore')

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
d1 = pd.read_csv("./data/fraud/Fraud_Data.csv");
d2 = pd.read_csv("./data/fraud/IpAddress_to_Country.csv");
```

```python
#l'execution prend un peu de temps : 10 - 15 min

countries = []
for i in range(len(d1)):
    ip_address = d1.loc[i, 'ip_address']
    tmp = d2[(d2["lower_bound_ip_address"] <= ip_address) & (d2["upper_bound_ip_address"] >= ip_address)]
    
    if len(tmp) == 1:
        countries.append(tmp["country"].values[0])
    else:
        countries.append('NA')
        
d1["country"] = countries
d1
```

```python
d1['signup_time'] = pd.to_datetime(d1['signup_time'])
d1['purchase_time'] = pd.to_datetime(d1['purchase_time'])
```

```python
time_diff = d1['purchase_time'] - d1['signup_time']
time_diff = time_diff.apply(lambda x: x.total_seconds())

d1['time_diff']=time_diff
```

```python
device_num = d1[['user_id', 'device_id']].groupby('device_id').count().reset_index()
 
device_num = device_num.rename(columns={'user_id': 'device_num'})
d1 = d1.merge(device_num, how='left', on='device_id')
```

```python
ip_num = d1[['user_id', 'ip_address']].groupby('ip_address').count().reset_index()
 
ip_num = ip_num.rename(columns={'user_id': 'ip_address_num'})
d1 = d1.merge(ip_num, how='left', on='ip_address')
```

```python
# Signup day and week
d1['signup_day'] = d1['signup_time'].apply(lambda x: x.dayofweek)
d1['signup_week'] = d1['signup_time'].apply(lambda x: x.week)
# Purchase day and week
d1['purchase_day'] = d1['purchase_time'].apply(lambda x: x.dayofweek)
d1['purchase_week'] = d1['purchase_time'].apply(lambda x: x.week)
```
